Compare feature information for matching arrivals between the catalog

```python
merge_arrs[merge_arrs['_merge'] == 'both']
```

```python
# Columns names so cols can be in the same order and remove arrival id
cols = ['event_identifier', 'network', 'station',
       'channel', 'location_code', 'source_latitude', 'source_longitude',
       'source_receiver_distance_km', 'source_receiver_back_azimuth',
       'source_depth_km', 'noise_variance', 'noise_minimum_value',
       'noise_maximum_value', 'signal_variance', 'signal_minimum_value',
       'signal_maximum_value', 'noise_dominant_frequency',
       'noise_dominant_amplitude', 'signal_dominant_frequency',
       'signal_dominant_amplitude', 'magnitude_type', 'magnitude',
       'avg_noise_1.00', 'avg_signal_1.00', 'avg_noise_2.00',
       'avg_signal_2.00', 'avg_noise_3.00', 'avg_signal_3.00',
       'avg_noise_4.00', 'avg_signal_4.00', 'avg_noise_5.00',
       'avg_signal_5.00', 'avg_noise_6.00', 'avg_signal_6.00',
       'avg_noise_7.00', 'avg_signal_7.00', 'avg_noise_8.00',
       'avg_signal_8.00', 'avg_noise_9.00', 'avg_signal_9.00',
       'avg_noise_10.00', 'avg_signal_10.00', 'avg_noise_11.00',
       'avg_signal_11.00', 'avg_noise_12.00', 'avg_signal_12.00',
       'avg_noise_13.00', 'avg_signal_13.00', 'avg_noise_14.00',
       'avg_signal_14.00', 'avg_noise_15.00', 'avg_signal_15.00',
       'avg_noise_16.00', 'avg_signal_16.00', 'avg_noise_17.00',
       'avg_signal_17.00', 'avg_noise_18.00', 'avg_signal_18.00']
```

```python
old_p_matching_feats_df = old_p_feats.merge(merge_arrs[merge_arrs['_merge'] == 'both'][['event_identifier', 'station', 'channel']], how='inner').sort_values(['event_identifier', 'station', 'channel']).reset_index(drop=True)[cols]
print(old_p_matching_feats_df.shape)
old_p_matching_feats_df.head()
```

```python
new_p_matching_feats_df = new_p_feats.merge(merge_arrs[merge_arrs['_merge'] == 'both'][['event_identifier', 'station', 'channel']], how='inner').sort_values(['event_identifier', 'station', 'channel']).reset_index(drop=True)[cols]
print(new_p_matching_feats_df.shape)
new_p_matching_feats_df.head()
```

```python
atol = 0.1
inds = []
for col in cols:
    s_new = new_p_matching_feats_df[col].values
    s_old = old_p_matching_feats_df[col].values
    arr_eq = np.array_equal(s_new, s_old)
    # print(col, arr_eq)
    if not arr_eq:
        print(col)
        print("\tAll close:", np.allclose(s_new, s_old, rtol=1e-2))
        print(f"\t{np.not_equal(s_new, s_old).sum()}/{len(s_new)} elements are different, {(abs(s_new-s_old) > atol).sum()} are > {atol}")
        print(f"\taverage difference: {np.mean(s_new-s_old):0.2e}, std: {np.std(s_new-s_old):0.2e}, abs max: {np.max(abs(s_new-s_old)):0.2e}")
        inds.append(np.where(abs(s_new-s_old) > atol)[0])
inds = np.unique(np.concatenate(inds))
print(f"{inds.shape[0]} with some difference > {atol}")
```

```python
print(f"Station breakdown of arrivals with some feature difference > {atol}:")
old_p_matching_feats_df.loc[inds][['station', 'channel']].value_counts()
```

```python
bad_evids = old_p_matching_feats_df.iloc[inds].event_identifier.value_counts().reset_index()
print(f"{len(bad_evids)} events with > {atol} difference in some feature")
```
