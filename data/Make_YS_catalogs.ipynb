```python
import pandas as pd
import numpy as np
from obspy.core.utcdatetime import UTCDateTime
```

# Got these catalogs from Ben's directory on chpc on 5/14/23

```python
# Combine current arrival catalogs (2012-2021) from Ben
entire_3c_df = pd.read_csv("catalogs/currentEarthquakeArrivalInformation3C.csv")
entire_1c_df = pd.read_csv("catalogs/currentEarthquakeArrivalInformation1C.csv")
historical_3c_df = pd.read_csv("catalogs/historicalEarthquakeArrivalInformation3C.csv")
historical_1c_df = pd.read_csv("catalogs/historicalEarthquakeArrivalInformation1C.csv")
entire_catalog_df = pd.concat([entire_3c_df[entire_1c_df.columns], entire_1c_df,
                             historical_3c_df[historical_1c_df.columns], historical_1c_df], axis=0).sort_values("origin_time")
```

```python
entire_catalog_df["utc"] = entire_catalog_df.apply(lambda x: UTCDateTime(x.origin_time), axis=1)
entire_catalog_df.head()
```

```python
keith_catalog = pd.read_csv("cat.yellow_long", sep="\s+")
keith_catalog.head()
```

```python
keith_catalog["utc"] = keith_catalog.apply(lambda x: UTCDateTime(f"{x.Date}T{x.Time}"), axis=1)
keith_catalog.head()
```

# Filter ben's catalog by location - there are 52185K events in the Yellowstone region 

```python
# Filter to just be in Yellowstone region, anything greater than 43/44 should be in Yellowstone
ys_df = entire_catalog_df[(entire_catalog_df["event_lat"] > 43.5) & (entire_catalog_df["event_lat"] < 46) &
                         (entire_catalog_df["event_lon"] > -112.6) & (entire_catalog_df["event_lon"] < -109)].sort_values("evid")
```

```python
# Reduce to Yellowstone events occuring from 2002 to 2022 June 30
start_time = UTCDateTime("2002-01-01")
end_time = UTCDateTime("2022-07-01")
ys_df_2002= ys_df[(ys_df["utc"] > start_time) & (ys_df["utc"] < end_time)].copy()
keith_df_2002= keith_catalog[(keith_catalog["utc"] > start_time) & (keith_catalog["utc"] < end_time)].copy()
```
