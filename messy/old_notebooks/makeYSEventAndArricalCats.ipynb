# Filter ben's catalog by location - there are ~18K events in the Yellowstone region

```python
ys_gain_df =  entire_gain_catalog_df[(entire_gain_catalog_df["event_lat"] > 44) & (entire_gain_catalog_df["event_lat"] < 45.167) &
                         (entire_gain_catalog_df["event_lon"] > -111.133) & (entire_gain_catalog_df["event_lon"] < -109.75)].sort_values("evid")
```

### The slightly longer catalog without gains and the shorter catalog with gains contain the same events over the same time period

```python
# Does return False if put in a different date
# Use .dt.date because the 'utc' info is being stored in the pandas df, 
# but np.datetime64 doesn't keep track of time scales/zone anymore
# So it won't let me compare datetime64[UTC] and datetime64
np.array_equal(ys_df[ys_df['utc'].dt.date < np.datetime64('2022-07-28')].drop_duplicates('evid')['evid'], 
               ys_gain_df.drop_duplicates('evid')['evid'])
```
